```python
import pandas as pd
from chembl_webresource_client.new_client import new_client

import numpy as np
from rdkit import Chem
from rdkit.Chem import Descriptors, Lipinski
```

```python
target = new_client.target
target = target.search('coronavirus')
target
```

```python
df = pd.DataFrame.from_dict(target)
```

```python
df
```

We will retrieve only bioactivity data for coronavirus 3C-like proteinase (CHEMBL3927) that has a standard value as IC50
 values in nM (nanomolar) unit. The lower the standard value, the more potent the drug is.

```python
selected_target = df.target_chembl_id[6]
selected_target
```

```python
activity =  new_client.activity
res = activity.filter(target_chembl_id = selected_target).filter(standard_type = 'IC50')
df = pd.DataFrame.from_dict(res)
```

```python
df
```

```python
df.columns
```

```python
df = df[['molecule_chembl_id','canonical_smiles','standard_value']]
```

# Handling Missing Values


```python
df.standard_value.dropna(inplace = True)
```

```python
def bioactivity_class(value):
  if float(value) >= 10000:
    return "inactive"
  elif float(value) <= 1000:
    return "active"
  else:
    return "intermediate"
```

```python
df['bioactivity'] = df['standard_value'].map(bioactivity_class)
```
